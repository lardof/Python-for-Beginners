```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(15,10)*100
y = np.random.rand(15,10)*100

x2 = np.random.rand(15,10)*50
y2 = np.random.rand(15,10)*50
```

```python
plt.scatter(x,y);
plt.scatter(x2,y2);
```

```python
#2.asama
```

```python
plt.scatter(x,y,color = 'cyan');
plt.scatter(x2,y2,color='black',alpha=0.5);
plt.xlabel('yatay eksen',size=15);
plt.ylabel('dusey eksen',size=15);
plt.title('baslik',size=22);
```

```python
#3.asama
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,4))
plt.scatter(x,y,color = 'cyan')
plt.scatter(x2,y2,color='black',alpha=0.5)
plt.xlabel('yatay eksen',size=15)
plt.ylabel('dusey eksen',size=15)
plt.title('baslik',size=22)
fig.text(1,0.5,      #buradaki 1 ve 0.7 konumlarii belirlemeye yariyor
         '''
Bunu yapmak icin 3 tane yan 
yana den den koymak gerekiyor 
bunalari yazmamin sebebi bunlarin 
assagida biraz yer kaplamasi gerekiyor.
''',fontfamily='serif' ,fontsize=15,)
plt.legend(labels=['mavi','siyah'],bbox_to_anchor=(1.0, 0.2));
#plt.legend(labels=['mavi','siyah'],loc='lower left');
```
