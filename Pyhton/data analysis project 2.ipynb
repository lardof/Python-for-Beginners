```python
sales_g = data.Global.sum()
sales_n = data["North America"].sum()
sales_e = data.Europe.sum()
sales_j = data.Japan.sum()
sales_r = data["Rest of World"].sum()

print("Global Satislar : {}".format(sales_g))
print("North America Satislar : {}".format(sales_n))
print("Europe Satislar : {}".format(sales_e))
print("Japan Satislar : {}".format(sales_j))
print("Rest of World Satislar : {}".format(sales_r))
```

```python
labels = data.describe().columns[1:].tolist()
labels
```

```python
values =[sales_n, sales_e, sales_j, sales_r, sales_g]
```

```python
plt.bar(labels, values,color ="b")
plt.xlabel('Region')
plt.ylabel('Billion')
plt.title('Global Sales By Region')




plt.show()
```

```python
labels = data.describe().columns[1:-1].tolist()
values =[sales_n, sales_e, sales_j, sales_r,]
plt.pie(values, labels = labels, autopct = "%1.1f%%")
plt.title('Global Sales By Region')


plt.show()
```

```python
sales_by_year = data.groupby("Year")[["North America", "Europe", "Japan", "Rest of World"]].sum()

fig, ax = plt.subplots()
ax.plot(sales_by_year.index, sales_by_year["North America"], label="North America")
ax.plot(sales_by_year.index, sales_by_year["Europe"], label="Europe")
ax.plot(sales_by_year.index, sales_by_year["Japan"], label="Japan")
ax.plot(sales_by_year.index, sales_by_year["Rest of World"], label="Rest of World")
ax.set_xlabel("Year")
ax.set_ylabel("Sales (millions)")
ax.set_title("Video Game Sales by Year")
ax.legend()
plt.show()
```
