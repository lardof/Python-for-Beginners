```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('trends.csv')
data.head(10)
```

```python
data.info()
```

```python
location = data['location'].value_counts()
year = data['year'].value_counts()
category = data['category'].value_counts()
rank = data['rank'].value_counts()
query = data['query'].value_counts()
```

```python
print('veri kumesi {} kolon ve {} satirdan olusmaktadir.'.format(data.shape[1],data.shape[0]))
print("location kolonunda {} farkli ulke var.".format(location.shape[0]))
print('category kolonunda {} farkli kategori var.'.format(category.shape[0]))
print('rank kolonunda {} farkli siralama var.'.format(rank.shape[0]))
print('year kolonunda {} farkli yil var.'.format(year.shape[0]))
print('query kolonunda {} farkli sorgu var.'.format(query.shape[0]))
```

```python
x = query.index
x = x[0:10]
y = query
y = y[0:10]

plt.bar(x, y)
```

```python
x = query.index
x = x[0:10]
y = query
y = y[0:10]

fig, ax=plt.subplots(1,1, figsize=(10,8))
plt.rcParams['axes.facecolor'] = 'yellow'


plt.xticks(rotation=(45))
plt.title('2001-2019 yillari arasinda en cok trend olan 10 sorgu')
plt.bar(x, y, width=0.2, color = 'black')
fig.text(0.68,0.74, 
'''
Paul Walker gibi bir 
ismin en basta olmasinin 
sebebnini o sene hayatini 
kahybetmesine baglamamiz 
cok normal.
''')
plt.show()
```

```python
dataTR = data.loc[data['location']=='Turkey']

location = dataTR['location'].value_counts()
year = dataTR['year'].value_counts()
category = dataTR['category'].value_counts()
rank = dataTR['rank'].value_counts()
query = dataTR['query'].value_counts()

print('veri kumesi {} kolon ve {} satirdan olusmaktadir.'.format(dataTR.shape[1],data.shape[0]))
print("location kolonunda {} farkli ulke var.".format(location.shape[0]))
print('category kolonunda {} farkli kategori var.'.format(category.shape[0]))
print('rank kolonunda {} farkli siralama var.'.format(rank.shape[0]))
print('year kolonunda {} farkli yil var.'.format(year.shape[0]))
print('query kolonunda {} farkli sorgu var.'.format(query.shape[0]))
```

```python
x = query.index
x = x[0:10]
y = query
y = y[0:10]

fig, ax=plt.subplots(1,1, figsize=(10,8))
plt.rcParams['axes.facecolor'] = 'yellow'


plt.xticks(rotation=(45))
plt.title('2001-2019 yillari arasinda en cok trend olan 10 sorgu')
plt.bar(x, y, width=0.2, color = 'black')
fig.text(0.64,0.75, 
'''
Kiralik Ask ve Depremin
en baslarda yer aldigini
gormemiz mummkun.Devaminda
ise bizleri asure ve gullac
gibi yemekler bekliyor.
''')
plt.show()
```
