```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
list1 = [10,20,50,100]
list2 = [5,8,9,50]
list3 = ["A","B","C","D"]
```

```python
fig, ax=plt.subplots(1,1,figsize = (8,5))  #grafik boyutu belirleme
plt.bar(list3,list1,color="red")
plt.bar(list3,list2,color="blue")
plt.title("Market", size = 0)
plt.xlabel("yatay", size = 15)
plt.ylabel("dusey", size = 15)
plt.legend(labels=["kirmizi","mavi"])
fig.text(0.78,0.65,"""
                  oguz mete kaya hazaretleri
                  grafik olusturmaya caslisiyor 
                  burda ne yazdiginin 
                  cok onemi yok
                  """ , fontsize=15)
plt.show()
```
