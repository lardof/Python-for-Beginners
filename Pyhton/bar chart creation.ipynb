```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
list1 = ["X","Y","Z","T"]
list2 = [10,20,15,5]

plt.bar(list1,list2)
plt.title("Mete")
plt.show() #en sonda olmak zorunda
```

```python
list1 = ["X","Y","Z","T"]
list2 = [10,20,15,5]

plt.bar(list1,list2,0.2)
plt.title("Mete")
plt.show()
```

```python
list1 = ["X","Y","Z","T"]
list2 = [10,20,15,5]

plt.barh(list1,list2,0.2)   #bar sonun h getirince yatay olur
plt.title("Mete")
plt.show()
```

```python
list1 = ["X","Y","Z","T"]
list2 = [10,20,15,5]
list3 = [20,40,30,10]

plt.bar(list1,list3,width = 0.3, color = "black")
plt.bar(list1,list2,width = 0.3 , color = "red")

plt.title("Mete")
plt.xlabel("yatay")
plt.ylabel("dusey")
plt.legend(labels = ["siyah","kirmizi"])
plt.show()
```
